## Sentiment Analysis

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters

prmStatement = ["This is really disappointing; nothing went right.",
                "Today is the best day I've ever had!",
                "I'm having a hard time dealing with stress.",
                "I'm very pleased with how everything turned out.",
                "I'm somewhat let down by the outcome.",
                "I'm content with the results; they lived up to my expectations.",
                "It was decent—neither outstanding nor terrible.",
                "Everything is a total mess; I'm heartbroken.",
                "I can't believe how bad this experience was.",
                "I'm quite unhappy with how things ended up.",
                "This didn't turn out as well as I had hoped."]
# Predict
mdlSnt(prmStatement)
```

Can perform sentiment analysis in other languages as well using `model = 'nlptown/bert-base-multilingual-uncased-sentiment'`

```python
# Model
mdlSnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmStatement = "verso l'ignoto, vi ritroverete"  # bilinmeyene doğru, kendini bulacaksın

# Predict
mdlSnt(prmStatement)
```

The result is in stars as opposed to score... this is as per the model definition.  See model details in HF for more details.

## Question Answering

### Example

```python
from transformers import pipeline

# Initialize the question-answering pipeline
mdlQa = pipeline("question-answering")

# Defence on American Football
f = """In American football, the mentality of a defensive player revolves around several key aspects:

Defensive players must be highly alert and reactive, as they need to anticipate the offensive team's moves and adjust their strategies quickly. They must possess a strong understanding of the game’s tactics, including recognizing offensive patterns and adapting to various plays. This requires a combination of strategic thinking and quick decision-making.

A defensive player also needs to be mentally resilient and focused. They must stay composed under pressure, especially when faced with fast-paced and unpredictable offensive attacks. This involves maintaining concentration throughout the game and not letting mistakes or setbacks affect their performance.

Furthermore, a defensive player must have a strong sense of teamwork and communication. Coordinating with teammates to execute defensive schemes effectively is crucial. This includes understanding one's role within the defensive strategy and working cohesively with others to achieve a common goal.

In summary, the mentality for playing defense in American football involves being alert, adaptable, resilient, and communicative, with a strong focus on strategy and teamwork."""

# Questions related to powerlifting
question1 = "How does the ability to anticipate and react quickly influence a defensive player's performance on the field?"
question2 = "In what ways does mental resilience and focus contribute to a defensive player’s effectiveness during high-pressure situations?"
question3 = "Why is effective communication and teamwork crucial for defensive players, and how does it impact the overall success of a defensive strategy?"

# Predict
print("Answer 1:", mdlQa(question=question1, context=f))
print("Answer 2:", mdlQa(question=question2, context=f))
print("Answer 3:", mdlQa(question=question3, context=f))
```
